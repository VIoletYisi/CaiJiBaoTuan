#### Question 1

```python
# Define the Gaussian curve.
def f(x):
    return 1 / np.sqrt(2 * np.pi) * np.exp(- (x ** 2) / 2)
```

```python
# Plot thg Gaussian curve.
x = np.arange(-100,100,0.01)
y = f(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gaussian curve')
plt.show()
```

```python
# Define the Simpson's method, with N = 1001.
def Simpson(a,b,N=1001):
    def f(x):
        return 1 / np.sqrt(2 * np.pi) * np.exp(- (x ** 2) / 2)
    h = (b - a) / (N - 1)
    w = np.zeros((N-1,))
    w [0] = 1 / 3 * h
    w [-1] = 1 / 3 * h
    for i in range (1, N-1,2):
        w[i] =  4 / 3 * h
    for i in range(2, N-1, 2):
        w[i] = 2 / 3 * h
    I = 0
    for i in range(N-1):
        I += w[i] * f(a + i * h)
    return I
```

```python
Simpson(-0.5,0.5)
```

```python
Simpson(-1,1)
```
