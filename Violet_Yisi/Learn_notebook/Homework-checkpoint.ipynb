### Problem 3
#### Question (a)

```python
n = np.array([3, 4, 5, 100, 10000, 1000000])
p = np.sin(math.pi / n) * n
print(p)
print('n\tp\t\t\t' + 'n p')
for j in range(3):
    print('{:}\t{:.11f}'.format(n[j],p[j]) + '{:>12} {:<.11f}'.format(n[j + 3],p[j + 3]))
```

#### Question (b)

```python
dataframe = pd.DataFrame({'n':n,'p':p})
dataframe.to_csv('n_p.csv', index = False,sep=',')

#load from n_p.csv flie
print("\nLoad from csv:\n{}".format(np.genfromtxt("n_p.csv",delimiter=','))) 
```

#### Question (c)

```python
np.savez('np_values', n = n , p = p )

#load from np_values.npz file
with np.load('np_values.npz') as n_p: 
    print('\nLoad from npz:')
    print("n = {}".format(n_p["n"]))
    print("p = {}\n".format(n_p["p"]))
```

```python
plt.semilogx(n,p)
plt.title('semi-log plot of $p$ versus $n$')
plt.xlabel('$n$')
plt.ylabel('$p$')
plt.legend('p')
plt.show()
```

From the plot we are easy to know that when n is large enough, the value of p is going to be very close to PI. 

That is to say $\lim_{n \to \infty}{p}=\pi$

#### Question(b)

```python
v_average = np.diff(y) / np.diff(t)
print(v_average)
```

#### Question (c)

```python
t_mid=np.zeros(19,)
for i in range(19):
    t_mid[i] = (t[i] + t[i + 1]) / 2
a = np.diff(v_average) / np.diff(t_mid)
print(a)
```

#### Question (d)

```python
plt.plot(t[1:],v_average,'ro-', markevery=1)
plt.title('plot of $v$ versus $t$')
plt.xlabel('$t /s$',fontsize=14)
plt.ylabel('$v /(m/s)$',fontsize=14)
plt.legend('v')
plt.show()
```

```python
plt.plot(t_mid[1:],a,'go-', markevery=1)
plt.axis('equal')
plt.title('plot of $a$ versus $t_mid$')
plt.xlabel('$t /s$',fontsize=14)
plt.ylabel('$a /(m/s^2)$',fontsize=14)
plt.legend('a')
plt.show()
```

$\ \ \ \ $ Look at the plot, we can find out that velocity and time are not linearly dependent. However, the accerelation is constant, which means v-t should be linear.

$\ \ \ \ $ I think that's because the velocity that I pick is the average of the interval, but the time that I pick is the starting point of the interval.

$\ \ \ \ $ In order to eliminate this error, I suggest using t_mid instead of b.


    For the acceleration, because it's a constant, it doesn't have this problem, but for consistency, I would recommend t_mmid instead of t_mid



```python
t_mmid=np.zeros(18,)
for i in range(18):
    t_mmid[i] = (t_mid[i] + t_mid[i + 1]) / 2
```

```python
plt.plot(t_mid,v_average,'ro-', markevery=1)
plt.title('plot of $v$ versus $t_mid$')
plt.xlabel('$t /s$',fontsize=14)
plt.ylabel('$v /(m/s)$',fontsize=14)
plt.legend('v')
plt.show()
```

```python
plt.plot(t_mmid,a,'go-', markevery=1)
plt.axis('equal')
plt.title('plot of $a$ versus $t_mmid$')
plt.xlabel('$t /s$',fontsize=14)
plt.ylabel('$a /(m/s^2)$',fontsize=14)
plt.legend('a')
plt.show()
```
