<font size = 4 color = black face = "Times New Roman">Easily to find that the roots are in the region (-10.0, -8.0),(-8.0,-7.1),(-2.0,-0.001).But, there is a problem to calculate the root in reginon (-8.0, -7.1), Because the function goes to infinity here, which is very to calculate. Then we will use the sacnt method to get them.

```python
# Its a function that use sacnt method to get the root of a given function.
def secant(f, x2, x1, eps):
    f_x2 = f(x2)
    f_x1 = f(x1)
    iteration_counter = 0
    while abs(f_x2) > eps and iteration_counter < 100:
        try:
            denominator = (f_x1 - f_x2)/(x1 - x2)
            x = x1 - f_x1/denominator
        except ZeroDivisionError:
            print('Error! - denominator zero for x = ', x2)
            sys.exit(1)
        x2 = x1
        x1 = x
        f_x2 = f(x2)
        f_x1 = f(x1)
        iteration_counter = iteration_counter + 1
    if abs(f_x1) > eps:
        iteration_counter = -1
    return x
```

```python
# We define a function to get the ground state energy.
def energy(v,eps):
    def f(e):
        return np.sqrt(v + e) * np.tan(np.sqrt(v + e)) - np.sqrt(-e)
    x2 = -v
    f_x2 = f(x2)
    x1 = x2 + 0.001
    f_x1 = f(x1)
    while f_x2 * f_x1 > 0:
        x1 += 0.001
        f_x1 =f(x1)  
    iteration_counter = 0
    while abs(f_x2) > eps and iteration_counter < 100:
        try:
            denominator = (f_x1 - f_x2)/(x1 - x2)
            x = x1 - f_x1/denominator
        except ZeroDivisionError:
            print('Error! - denominator zero for x = ', x2)
            sys.exit(1)
        x2 = x1
        x1 = x
        f_x2 = f(x2)
        f_x1 = f(x1)
        iteration_counter = iteration_counter + 1
    if abs(f_x1) > eps:
        iteration_counter = -1
    return x
```

```python
v = np.arange(0.1,100,0.1)
e = np.arange(0.1,100,0.1)
l = v.size
for i in range(0,l):
    e[i] = energy(v[i],10**(-10))
plt.plot(e, v)
plt.xlabel('ground state energy $E$')
plt.ylabel('depth of the potential $V_0$')
plt.title('The ground state energy versus the depth of the potential')
plt.ylim(0, 100)
plt.show()
```

<font size = 4 color = black face = "Times New Roman">We make the diagram of the ground state energy versus the depth of the potential.It can be seen from the figure that the ground state energy is negatively correlated with the depth of the potential well. The deeper the potential well is, the smaller the ground state energy is.
