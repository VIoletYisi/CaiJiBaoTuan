```python
import numpy as np
from scipy.special import *
```

<font size = 5 color = black face = "Times New Roman">Firstly we should get True Values, Up_method Values and Down_method Values.

```python
tv = [0.1 , 1 , 10] #values used to test my program
```

```python
# This part aims to get the true value of the problem.
j = np.empty((26,3))
for i in range(26):
    for k in range(3):
        j[i][k] = spherical_jn(i,tv[k])
print('True Values')
print('l\t\t\t0.1\t\t\t\t\t1\t\t\t\t\t10')
print('------------------------------------------------------------------------------------------------------------------------------------')
for i in range(26):
    print('{}\t\t\t{:<20}\t\t\t{:<20}\t\t\t{:<20}'.format(i,j[i][0],j[i][1],j[i][2]))
```

### Method Upward


```python
# Get datas by using upward method
j_u = np.empty((26,3))
for i in range (3):
    j_u[0][i] = spherical_jn(0,tv[i])
    j_u[1][i] = spherical_jn(1,tv[i])
for i in range(1,25):
    for k in range(3):
        j_u[i + 1][k] = (2 * i + 1) / tv[k] * j_u[i][k] - j_u[i-1][k] 
print('The values what we get by using upward method')
print('l\t\t\t0.1\t\t\t\t\t1\t\t\t\t\t10')
print('------------------------------------------------------------------------------------------------------------------------------------')
for i in range(26):
    print('{}\t\t\t{:<20}\t\t\t{:<20}\t\t\t{:<20}'.format(i,j_u[i][0],j_u[i][1],j_u[i][2]))
```

```python
# Relative error of up method
j_u_error = np.empty((26,3))
for i in range(26):
    for k in range(3):
        j_u_error[i][k] = abs(j[i][k] - j_u[i][k]) / abs(j[i][k])
print('The relative error of upward method')
print('l\t\t\t0.1\t\t\t\t\t1\t\t\t\t\t10')
print('------------------------------------------------------------------------------------------------------------------------------------')
for i in range(26):
    print('{}\t\t\t{:<20}\t\t\t{:<20}\t\t\t{:<20}'.format(i,j_u_error[i][0],j_u_error[i][1],j_u_error[i][2]))
```

### Method Downward

```python
# Get datas by using upward method
j_d = np.empty((26,3))
for i in range (3):
    j_d[25][i] = spherical_jn(25,tv[i])
    j_d[24][i] = spherical_jn(24,tv[i])
for i in range(24,0,-1):
    for k in range(3):
        j_d[i - 1][k] = (2 * i + 1) / tv[k] * j_d[i][k] - j_d[i + 1][k] 
print('The values what we get by using downwar method')
print('l\t\t\t0.1\t\t\t\t\t1\t\t\t\t\t10')
print('------------------------------------------------------------------------------------------------------------------------------------')
for i in range(26):
    print('{}\t\t\t{:<20}\t\t\t{:<20}\t\t\t{:<20}'.format(i,j_d[i][0],j_d[i][1],j_d[i][2]))
```

```python
# Relative error of down method
j_d_error = np.empty((26,3))
for i in range(26):
    for k in range(3):
        j_d_error[i][k] = abs(j[i][k] - j_d[i][k]) / abs(j[i][k])
print('The relative error of downward method')
print('l\t\t\t0.1\t\t\t\t\t1\t\t\t\t\t10')
print('------------------------------------------------------------------------------------------------------------------------------------')
for i in range(26):
    print('{}\t\t\t{:<20}\t\t\t{:<20}\t\t\t{:<20}'.format(i,j_d_error[i][0],j_d_error[i][1],j_d_error[i][2]))
```

### <font size = 5 color = black face = "Times New Roman">Question a)

<font size = 5 color = black face = "Times New Roman">Show the convergence and stability of your results

<font size = 5 color = black face = "Times New Roman">From the results of the above program, we can see that for upward method, the result is not convergent, and its stability is poor. What`s more, its relative error is so large that makes us  have enough reasons to believe that using this method is a bad decision. As opposed to this, downward method has a very satisfactory performance. Its result is convergent and stable. At the same time, its relative error is even smaller than $10^{-10}$. Needless to say, downward method is an excellent method.

### <font size = 5 color = black face = "Times New Roman">Question b)
